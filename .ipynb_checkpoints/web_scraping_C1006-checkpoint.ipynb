```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
pag_web = requests.get ('https://es.wikipedia.org/wiki/Anexo:Tabla_estad%C3%ADstica_de_la_Copa_Mundial_de_F%C3%BAtbol#cite_note-1')

# lo convertimos a un archivo soup

sopa = bs(pag_web.content)

# extraemos el html

contenido = sopa.prettify()

print (contenido)
```

```python
box_simbologia = sopa.find(class_='columns')
print (box_simbologia.prettify())
```

```python
fila_simbologia = box_simbologia.find_all('li')

for fila in fila_simbologia:
    print (fila.prettify())
```

```python
for index, fila in enumerate(fila_simbologia):
    if index == 0:
        print (fila.prettify())
```

```python
def extrae_lista(fila_dato):
    if fila_dato.find("li"):
        return [li.get_text() for li in fila_dato.find_all('li')]
    else:
        return fila_dato.find('td').get_text()

    
```

```python
referencias = {}


for index, fila in enumerate(fila_simbologia):
    referencias[index] = fila.get_text()
    
```
